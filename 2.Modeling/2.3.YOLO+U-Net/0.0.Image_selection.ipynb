```python
gdf_all = gpd.read_file("/shared/data/climateplus2025/Prediction_for_poster_July21/0.Image_files_selection/final_annotations.gpkg")
gdf_all.head(5)
```

```python
# from 5K dataset
excluded_images = ['2023_RGB_8cm_W57B_10', '2023_RGB_8cm_W57B_4', '2023_RGB_8cm_W57B_15',
 '2023_RGB_8cm_W53B_24', '2023_RGB_8cm_W36B_15', '2023_RGB_8cm_W16C_20',
 '2023_RGB_8cm_W57B_13', '2023_RGB_8cm_W53B_9', '2023_RGB_8cm_W25A_9',
 '2023_RGB_8cm_W17B_3', '2023_RGB_8cm_W36B_20', '2023_RGB_8cm_W17B_11',
 '2023_RGB_8cm_W57C_5', '2023_RGB_8cm_W17B_6', '2023_RGB_8cm_W17B_7']

gdf_filtered = gdf_all[~gdf_all['image_name'].isin(excluded_images)]

# Check result
print(gdf_filtered['image_name'].unique())
gdf_filtered.head(5)
```

```python
gdf_filtered['PV_normal_bin'] = gdf_filtered['PV_normal'].notna().astype(int)
gdf_filtered['PV_heater_bin'] = gdf_filtered['PV_heater'].notna().astype(int)
gdf_filtered['PV_pool_bin'] = gdf_filtered['PV_pool'].notna().astype(int)
gdf_filtered['uncertflag_bin'] = gdf_filtered['uncertflag'].notna().astype(int)

summary_df = gdf_filtered.groupby('image_name')[[
    'PV_normal_bin', 'PV_heater_bin', 'PV_pool_bin', 'uncertflag_bin'
]].sum().astype(int)

print(summary_df.head())

summary_df['total_pv'] = summary_df['PV_normal_bin'] + summary_df['PV_heater_bin'] + summary_df['PV_pool_bin']

top5_images = summary_df.sort_values('total_pv', ascending=False).head(10)

# Chage to top 5 to top 100
print(top5_images[['PV_normal_bin', 'PV_heater_bin', 'PV_pool_bin', 'total_pv']])
```
